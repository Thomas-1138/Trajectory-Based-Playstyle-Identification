```python
# Function to read .xz archives from ESTA
def read_parsed_demo(filename):
    with lzma.LZMAFile(filename, "rb") as f:
        d = json.load(f)
        return d
```

```python
# Function to create a vector representation for a game state (also called a frame)
def create_frame_row(frame):
    frame_data = {}

  # Global info
    frame_data["secondsSincePhaseStart"] = frame["seconds"]
    frame_data["bombPlanted"] = frame["bombPlanted"]
    frame_data["bombsite"] = frame["bombsite"]
    frame_data["totalSmokes"] = len(frame["smokes"])
    frame_data["totalFires"] = len(frame["fires"])

    # Team specific info (CT)
    frame_data["ctAlive"] = 0
    frame_data["ctHp"] = 0
    frame_data["ctArmor"] = 0
    frame_data["ctHelmet"] = 0
    frame_data["ctEq"] = 0
    frame_data["ctUtility"] = 0
    frame_data["ctEqValStart"] = 0
    frame_data["ctBombZone"] = 0
    frame_data["defusers"] = 0
    for p in frame["ct"]["players"]:
        frame_data["ctEqValStart"] += p["equipmentValueFreezetimeEnd"]
        if p["isAlive"]:
            frame_data["ctAlive"] += 1
            frame_data["ctHp"] += p["hp"]
            frame_data["ctArmor"] += p["armor"]
            frame_data["ctHelmet"] += p["hasHelmet"]
            frame_data["ctEq"] += p["equipmentValue"]
            frame_data["ctUtility"] += p["totalUtility"]
            frame_data["defusers"] += p["hasDefuse"]
            if p["isInBombZone"]:
                frame_data["ctBombZone"] += 1
  
  # Team specific info (T)
    frame_data["tAlive"] = 0
    frame_data["tHp"] = 0
    frame_data["tArmor"] = 0
    frame_data["tHelmet"] = 0
    frame_data["tEq"] = 0
    frame_data["tUtility"] = 0
    frame_data["tEqValStart"] = 0
    frame_data["tHoldingBomb"] = 0
    frame_data["tBombZone"] = 0
    for p in frame["t"]["players"]:
        frame_data["tEqValStart"] += p["equipmentValueFreezetimeEnd"]
        if p["isAlive"]:
            frame_data["tAlive"] += 1
            frame_data["tHp"] += p["hp"]
            frame_data["tArmor"] += p["armor"]
            frame_data["tHelmet"] += p["hasHelmet"]
            frame_data["tEq"] += p["equipmentValue"]
            frame_data["tUtility"] += p["totalUtility"]
            if p["isInBombZone"]:
                frame_data["tBombZone"] += 1
            if p["hasBomb"]:
                frame_data["tHoldingBomb"] = 1

    return frame_data
```

```python
# Function to create a set representation of a game state. Each element in the set is a vector of [global info, player-specific info]
def create_frame_set(frame, map_name):
  # Global info
    seconds = frame["seconds"]
    bomb_planted = 0
    if frame["bombPlanted"]:
        bomb_planted = 1
    bombsite_none = 0
    bombsite_a = 0
    bombsite_b = 0
    if frame["bombsite"] == "":
        bombsite_none = 1
    if frame["bombsite"] == "A":
        bombsite_a = 1
    if frame["bombsite"] == "B":
        bombsite_b = 1

    total_smokes = len(frame["smokes"])
    total_fires = len(frame["fires"])

  # Player specific info
    player_list_ct = []
    velocity_list_ct = []
    for p in frame["ct"]["players"]:
        helmet = 0
        if p["hasHelmet"]:
            helmet = 1
        kit = 0
        if p["hasDefuse"]:
            kit = 1
        bomb = 0
        if p["hasBomb"]:
            bomb = 1
        in_bomb_zone = 0
        if p["isInBombZone"]:
            in_bomb_zone = 1

        has_rifle = 0
        has_smg = 0
        has_heavy = 0
        if p["inventory"]:
            for w in p["inventory"]:
                if w["weaponClass"] == "Rifle":
                    has_rifle = 1
                if w["weaponClass"] == "SMG":
                    has_smg = 1
                if w["weaponClass"] == "Heavy":
                    has_heavy = 1

        player_vec = [
                      #seconds,
                      p["x"],
                      p["y"],
                      p["z"],
                      #p["velocityX"],
                      #p["velocityY"],
                      #p["velocityZ"],
                      #p["viewX"],
                      #p["viewY"],
                      #1, # 1 for CT, 0 for T

        ]
        velocity_vec = [
                      #seconds,
                      #p["x"],
                      #p["y"],
                      #p["z"],
                      p["velocityX"],
                      p["velocityY"],
                      p["velocityZ"],
                      #p["viewX"],
                      #p["viewY"],
                      #1, # 1 for CT, 0 for T

        ]
        player_list_ct.append(np.array(player_vec))
        velocity_list_ct.append(np.array(velocity_vec))
        
    player_list_t = []
    velocity_list_t = []
    for p in frame["t"]["players"]:
        helmet = 0
        if p["hasHelmet"]:
            helmet = 1
        kit = 0
        if p["hasDefuse"]:
            kit = 1
        bomb = 0
        if p["hasBomb"]:
            bomb = 1
        in_bomb_zone = 0
        if p["isInBombZone"]:
            in_bomb_zone = 1

        has_rifle = 0
        has_smg = 0
        has_heavy = 0
        if p["inventory"]:
            for w in p["inventory"]:
                if w["weaponClass"] == "Rifle":
                    has_rifle = 1
                if w["weaponClass"] == "SMG":
                    has_smg = 1
                if w["weaponClass"] == "Heavy":
                    has_heavy = 1

        player_vec = [
                      #seconds,
                      p["x"],
                      p["y"],
                      p["z"],
                      #p["velocityX"],
                      #p["velocityY"],
                      #p["velocityZ"],
                      #p["viewX"],
                      #p["viewY"],
                      #0, # 1 for CT, 0 for T

        ]
        velocity_vec = [
                      #seconds,
                      #p["x"],
                      #p["y"],
                      #p["z"],
                      p["velocityX"],
                      p["velocityY"],
                      p["velocityZ"],
                      #p["viewX"],
                      #p["viewY"],
                      #0, # 1 for CT, 0 for T

        ]
        player_list_t.append(np.array(player_vec))
        velocity_list_t.append(np.array(velocity_vec))
    return np.array(player_list_ct), np.array(velocity_list_ct), np.array(player_list_t), np.array(velocity_list_t)
```

## Total Distance

```python
def add_total_distances(df):
    total_distances = []
    for i in range(df.shape[0]):
        dist = 0
        for j in range(df.shape[1]-1):
            try:            
                point1 = df.iloc[i,j]
                point2 = df.iloc[i,j+1]
                # remember this is the distance in 3D and NOT 2D
                dist += math.dist(point1, point2)
            except:
                pass
        total_distances.append(dist)
    return total_distances
```

## Start to End Distance

```python
def add_start_to_end_distances(df):
    ste_distances = []
    for i in range(df.shape[0]):
        for j in range(1, df.shape[1]+1):
            try:
                ste_distance = math.dist(df.iloc[i,0], df.iloc[i,-j])
                ste_distances.append(ste_distance)
                break
            except:
                pass
    return ste_distances
```

```python
def angle(df, min_angle):
    """Return the angle between vectors
    """
    angles = []
    for i in range(df.shape[0]):
        #angle = 0
        angle = []
        for j in range(df.shape[1]-2):
            try:
                # calculate the angle between the current point and the next poin
                angle1 = np.arctan2(df.iloc[i][j+1][1] - df.iloc[i][j][1], df.iloc[i][j+1][0] - df.iloc[i][j][0])
                # calculate the angle between the next point and the point after that
                angle2 = np.arctan2(df.iloc[i][j+2][1] - df.iloc[i][j+1][1], df.iloc[i][j+2][0] - df.iloc[i][j+1][0])              
                # calculate the angle between the three consecutive points
                angle = angle2 - angle1
                print(angle)
                # add the angle to the list
                #if np.degrees(angle) > min_angle:
                    #angle += 1
                angle.append(np.degrees(angle))
            except:
                pass
                #print(j)
            angles.append(angle)
    return angles
```

```python
def angle_test(df, min_angle):
    angles = []
    for i in range(df.shape[0]):
        #angle = 0
        angle = []
        for j in range(df.shape[1]-2):
            try:   
                ba = df.iloc[i,j][0] - df.iloc[i,j+1][0], df.iloc[i,j][1] - df.iloc[i,j+1][1]
                bc = df.iloc[i,j+2][0] - df.iloc[i,j+1][0], df.iloc[i,j+2][1] - df.iloc[i,j+1][1]
                cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
                ang = np.arccos(cosine_angle)
                #print(np.degrees(ang))
                if (0 < np.degrees(ang) < 180-min_angle) or (180+min_angle < np.degrees(ang) < 360):
                    angle.append(ang)
            except:
                break
        angles.append(len(angle))
    return angles
```

## Average Speed

```python
def avg_speed(df):
    avg_speed = []
    for i in range(df.shape[0]):
        x,y,z = 0,0,0
        error = [df.shape[1]]
        for j in range(df.shape[1]):
            try:
                x +=  abs(df.iloc[i,j][0])
                y += abs(df.iloc[i,j][1])
                z +=  abs(df.iloc[i,j][2])
            except:
                error.append(j+1)
                break
        tmp = [x/error[0],y/error[0],z/error[0]]
        avg_speed.append(tmp)
    return avg_speed
```

# Convex Hull (covered area)

```python
convex_hull = np.zeros((5))
hull_centroids = np.zeros((5,2))
for i in range(len(telemetry_c)):
    hull = ConvexHull(np.array(telemetry_c[i]))
    convex_hull[i] = hull.volume
    hull_centroids[i] = (np.mean(hull.points[hull.vertices,0]), np.mean(hull.points[hull.vertices,1]))
convex_hull
```
